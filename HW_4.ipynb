Broadcasting refers to the automatic expansion of one or more arrays to make a computation "make sense." Here's a simple example: 

```python
a = np.array([0, 1, 2])
b = np.ones((3, 3))
a.shape, b.shape
```
```
((3,), (3, 3))
```
```python
a+b
```
```
array([[1., 2., 3.],
       [1., 2., 3.],
       [1., 2., 3.]])
```

What has happened here is that the first array `a` has been "broadcast" from a 1d array into a 2d array of size 3x3 in order to match the dimensions of `b`. The broadcast version of `a` looks like this: 

```
array([[0., 1., 2.],
       [0., 1., 2.],
       [0., 1., 2.]])
```

Consult the notes above for many more examples of broadcasting. Pay special attention to the discussion of the [rules of broadcasting](https://jakevdp.github.io/PythonDataScienceHandbook/02.05-computation-on-arrays-broadcasting.html#Rules-of-Broadcasting).

### (C). 

Implement an automated test class called `TestBroadcastingRules` which tests the three rules of array broadcasting. 

> **Rule 1**: If the two arrays differ in their number of dimensions, the shape of the one with fewer dimensions is padded with ones on its leading (left) side.

To test this rule, write a method `test_rule_1()` that constructs the arrays:

> **Rule 3**: If in any dimension the sizes disagree and neither is equal to 1, an error is raised.

To test this rule, write a method `test_rule_3` that constructs the arrays 

```

b = np.ones((3, 3))
```

It should then attempt to construct `c = a + b`. The test should *pass* if the Rule 3 error is raised, and fail otherwise. You will need to figure out what kind of error is raised by Rule 3 (is it a `TypeError`? `ValueError`? `KeyError`?). Your tests have passed if no output is printed when you run this code. 

### Your Solution

````python
# write your tester class here
import unittest

class TestBroadcastingRules(unittest.TestCase):
    def test_rule_1(self):
        '''a has been broadcasted to fit the shape of b. The new shape is a 2d 1x3 array. By rule 1, a has 
        fewer dimensions, so it is padded to the left with 1 to fit the shape of b (1,3). The last value for
        c is 3. because the last column for a and b (1 and 2) add to 3.'''
        a = np.ones(3)
        b = np.arange(3).reshape(1, 3) 
        c = a + b
        self.assertEqual(c.shape,(1,3))
        self.assertEqual(c[-1][-1],3.)
    
    def test_rule_2(self):
        ''' a has been broadcasted to fit the shape of b. The new shape is a 2d 3x3 array. By rule 2, a has
        shape 1 so it must be stretched to fit the shape of b. The value for c[1,2] is 6. because
        the corresponding index for a is 1. and for b is 5. This adds to 6'''
        a = np.ones((1, 3))
        b = np.arange(9).reshape(3, 3)
        c = a + b
        self.assertEqual(c.shape,(3,3))
        self.assertEqual(c[1,2],6.0)

    def test_rule_3(self):
        '''According to rule 3, an error is raised because a and b do not have matching dimensions
        and neither shapes have shape 1, so they cannot be stretched to fit the other. Thus,
        they can not be added.'''
        a = np.ones((2, 3))
        b = np.ones((3, 3))
        with self.assertRaises(ValueError):
            c=a+b

````

````python
# run your tests
# your tests have passed if no output or errors are shown. 

tester = TestBroadcastingRules()
tester.test_rule_1()
tester.test_rule_2()
tester.test_rule_3()
````

Recall the simple random walk.  At each step, we flip a fair coin. If heads, we move "foward" one unit; if tails, we move "backward." 

## (A).

Way back in Homework 1, you wrote some code to simulate a random walk in Python. 

Start with this code, or use posted solutions for HW1. If you have since written random walk code that you prefer, you can use this instead. Regardless, take your code, modify it, and enclose it in a function `rw()`. This function should accept a single argument `n`, the length of the walk. The output should be a list giving the position of the random walker, starting with the position after the first step. For example, 

```python
rw(5)
[1, 2, 3, 2, 3]
```

Unlike in the HW1 problem, you should not use upper or lower bounds. The walk should always run for as long as the user-specified number of steps `n`. 

Use your function to print out the positions of a random walk of length `n = 10`. 

Don't forget a helpful docstring! 

````python
# solution (with demonstration) here
import random
def rw(n):
    
    '''This function takes in an integer n which signifies the length of the walk. The 
    function then randomly produces a choice of either going forward or backwards n times.
    The position will either increase or decrease by 1 depending on the choice.
    A list of length n (the number of steps) of the positions is returned'''
    results=[] #keeps track of forwards,backwards
    positions=[] #keeps track of number position
    counter=0     #position number
    for i in range(n):
        x = random.choice(["forwards","backwards"])
        results.append(x)
        if x=="forwards":        #increase position number, if going forward
            counter+=1
            positions.append(counter)
        elif x=="backwards":          #decrease position number, if going backwards
            counter-=1
            positions.append(counter)
    return(positions)
rw(10)
````
