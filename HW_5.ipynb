## Problem 1: Faceted Histogram

Run the following code block to define a function which generates two 1-dimensional `numpy` arrays. The first array, called `groups`, consists of integers between `0` and `n_groups - 1`, inclusive. The second array, called `data`, consists of real numbers. 

````python
import numpy as np
from matplotlib import pyplot as plt

def create_data(n, n_groups):
    """
    generate a set of fake data with group labels. 
    n data points and group labels are generated. 
    n_groups controls the number of distinct groups. 
    Returns an np.array() of integer group labels and an 
    np.array() of float data. 
    """
    
    # random group assignments as integers between 0 and n_groups-1, inclusive
    groups = np.random.randint(0, n_groups, n)
    
    # function of the groups plus gaussian noise (bell curve)
    data   = np.sin(groups) + np.random.randn(n)
    
    return(groups, data)
````

### Requirements: 

1. Your function should work **whenever `m_rows*m_cols` is equal to the total number of groups.** Your function should first check that this is the case, and raise an informative `ValueError` if not. You may assume that there is at least one data point for each group label in the data supplied. 
2. For full credit, you should not loop over the individual entries of `groups` or `data`. It is acceptable to loop over the distinct values of `groups`. In general, aim to minimize `for`-loops and maximize use of `Numpy` indexing. 
2. Use of `pandas` is acceptable but unnecessary, and is unlikely to make your solution significantly simpler. 
2. You should include a horizontal axis label (of your choice) along **only the bottom row** of axes. 
3. You should include a vertical axis label (e.g. "Frequency") along **only the leftmost column of axes.**

### Hints

- If your plots look "squished," then `plt.tight_layout()` is sometimes helpful. Just call it after constructing your figure, with no arguments. 
- Integer division `i // j` and remainders `i % j` are helpful here, although other solutions are also possible. 


````python
# your solution here
def facet_hist(groups,data,m_rows,m_cols,figsize,**kwargs):
    '''This function takes in 5 arguments. groups, is the array of group labels created by create_data,
    data is an array of data created by create_data, m_rows signifies the number of
    rows in the faceted histogram, m_cols signifies the number of columns, figsize signifies the figsize 
    of the plots. All of these arguments will be used to output a faceted histogram of m_rows and
    m_cols.'''
    if m_rows*m_cols!=(np.amax(groups)+1): #checking that rows x columns equals number of groups
        raise ValueError("Error!Product of rows and colums must equal number of groups.")
    else:
        fig,ax=plt.subplots(m_rows,m_cols,figsize=figsize)
        unique_groups=set(groups)
        for x in unique_groups:  #iterating through each distinct group label

            rows =x%m_rows    #finds index of rows and columns
            cols=x//m_rows
            
            ix=groups==x  #creates boolean array of when group array element 
                          #is the specific group label we are looking at
            group_name="Group " +str(x)
           
            ax[rows][cols].hist(data[ix],**kwargs) #plots corresponding data to group
            ax[rows][cols].set(title=group_name)
            
            if cols==0:
                ax[rows][cols].set(ylabel="Density") #left hand plots have y-label
            
            if rows==m_rows-1:
                ax[rows][cols].set(xlabel="x") #bottom-most plots have x-label
            
            
                                                      
        plt.tight_layout()    
            
            
     
        
````

Write a function called `scatterplot_matrix()`, with arguments `cols` and `figsize`. The `cols` argument should be a list of strings, each of which are the name of one of the columns above, for example `cols = ["Score", "GDP per capita", "Social support"]`. There is a separate scatterplot for each possible pair of variables. In fact, there are two: one where the first variable is on the horizontal axis, and one where it's on the vertical axis. Some analysts prefer to remove half the plots to avoid redundancy, but you don't have to bother with that. The diagonal is empty, since there's no point in investigating the relationship between a variable and itself. 

Don't forget comments and docstrings! 

````python
# define your function
def scatterplot_matrix(cols,figsize):
    '''This function takes in cols which is a list of strings signifying
    the column names of the happiness dataframe. figsize signifies the 
    figsize of the plots. The function uses these arguments to create a
    matrix of scatterplots that plots each possible pair of column 
    variables.'''
    fig,ax=plt.subplots(len(cols),len(cols),figsize=figsize) #creates matrix of subplots
    for x in range(len(cols)): #nested for loop that allows for each pair of column variables 
        for k in range(len(cols)): #to be plotted
            if x!=k:
                
                my_number=np.corrcoef(happiness[cols[x]].values,happiness[cols[k]].values)[0][1] #calculates
                                                                                                 #correlation coefficient
                ax[x][k].scatter(happiness[cols[x]],happiness[cols[k]],s=12) #plot the column variables
                ax[x][k].set(xlabel = r"$\rho$ = " + str(np.round(my_number, 2)),ylabel=cols[k],title=cols[x])
            else:
                ax[x][k].set(ylabel=cols[k],title=cols[x]) #if x and y axis are same variable, don't plot
    plt.tight_layout() 
        
````

````python
# test your code, several times if needed, and discuss the correlations you observe. 
# Add code cells if needed to show multiple outputs. 
cols = ["Score", 
        "GDP per capita", 
        "Social support"]
scatterplot_matrix(cols,figsize = (7,7))
````
